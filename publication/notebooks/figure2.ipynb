```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/pyTCR_benchmarks.csv')

# Remove the results from sample size of 2
df = df[df.Number_of_samples != 2]
df
```

## Figure2a

```python
sns.set_style("white") 
sns.set_context("talk") 

fig, ax = plt.subplots(figsize=(10, 10))
sns.lineplot(data=df, x="Number_of_samples", y="CPU_time", hue="Tool", style="Tool", markers=True)
ax.set_ylabel('CPU Time (s)')
ax.set_xlabel('Number of samples')
sns.despine()

plt.savefig("../figures/main/figure2a.pdf",bbox_inches="tight")
plt.savefig("../figures/main/figure2a.png",bbox_inches="tight")

plt.show()
```

## Figure2b

```python
sns.set_style("white") 
sns.set_context("talk") 

fig, ax = plt.subplots(figsize=(10, 10))
sns.lineplot(data=df, x="Number_of_samples", y="Memory", hue="Tool", style="Tool", markers=True)
ax.set_ylabel('Memory (MB)')
ax.set_xlabel('Number of samples')
sns.despine()

plt.savefig("../figures/main/figure2b.pdf",bbox_inches="tight")
plt.savefig("../figures/main/figure2b.png",bbox_inches="tight")

plt.show()
```
